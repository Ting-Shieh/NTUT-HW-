```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()

eta_list =np.arange(0.1,1.1,0.05)
overmath=False
for eta in eta_list:
    print('============ eta = {} ========'.format(eta))
    w = np.array([1,1.5]) # initial weights of x & y
    grad = np.array([2,2]) # gradient of x & y
    epislon = 0.001 # acceptable accuracy tolerance
    for k in range(10000):
        w_next =  grad * w # elementwise mpy
        if (np.sum(w_next**2) < epislon): #accuracy reached
            break
        w = w - eta * w_next
        #print(w, w_next)
        f_value = w[0]**2 + w[1]**2 + 3 # hard code x^2 + y^2 + 3
        if math.isinf(f_value):
            overmath=True
            
            break
        else:
            print(f_value)

    if not overmath:    
        if (abs(f_value-3.0)>epislon):
            print('Algorithm does not coverge!')
        else:

            df= df.append({'No of iter':k+1,'weights_left':w[0],'weights_right':w[1],'opt value':f_value, 'eta':eta},ignore_index=True)
            print('No of iter = ',(k+1),', weights =',w, ', opt value =', f_value)
            print(df)
    else:
        break
```
